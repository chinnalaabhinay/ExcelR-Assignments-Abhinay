```python
# ============================================================
# RANDOM FOREST, BAGGING & BOOSTING - GLASS DATASET
# Dataset Source: UCI Machine Learning Repository
# ============================================================

import warnings
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Optional: keep warnings visible but suppress specific noisy sklearn future warnings
# (comment out if you want to see all warnings)
warnings.filterwarnings("ignore", category=FutureWarning)

# ------------------------------------------------------------
# 1. LOAD DATA (UCI GLASS DATASET AUTOMATICALLY)
# ------------------------------------------------------------
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data"
columns = ["Id","RI","Na","Mg","Al","Si","K","Ca","Ba","Fe","Type"]
df = pd.read_csv(url, names=columns)

# Remove Id column
df.drop("Id", axis=1, inplace=True)

print("Dataset Loaded Successfully!")
print("Shape:", df.shape)
display(df.head())

# ------------------------------------------------------------
# 2. EXPLORATORY DATA ANALYSIS (EDA)
# ------------------------------------------------------------
print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
display(df.describe().T)

# Outlier detection (boxplots)
plt.figure(figsize=(12,6))
sns.boxplot(data=df.drop(columns=['Type']), orient='h')
plt.title("Boxplot of Features (Outlier Detection)")
plt.show()

# ------------------------------------------------------------
# 3. DATA VISUALIZATION
# ------------------------------------------------------------
# Histograms
df.hist(figsize=(12,10), bins=20)
plt.suptitle("Histogram of Features")
plt.show()

# Pairplot (small sample to keep it fast)
try:
    sns.pairplot(df.sample(min(150, len(df))), hue="Type", corner=True)
    plt.show()
except Exception as e:
    print("Pairplot skipped due to:", e)

# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.drop(columns=["Type"]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# ------------------------------------------------------------
# 4. DATA PREPROCESSING
# ------------------------------------------------------------
# Features & Target
X = df.drop("Type", axis=1)
y = df["Type"]

# Scaling (Standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# ------------------------------------------------------------
# HELPER FUNCTION FOR EVALUATION
# ------------------------------------------------------------
def evaluate(model_name, y_true, y_pred):
    """
    Evaluates and prints metrics. Uses zero_division=0 to avoid UndefinedMetricWarning
    when there are labels with no predicted samples.
    """
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
    rec = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)

    print(f"\n===== {model_name} =====")
    print("Accuracy :", round(acc,4))
    print("Precision:", round(prec,4))
    print("Recall   :", round(rec,4))
    print("F1-score :", round(f1,4))
    print("\nClassification Report:\n")
    print(classification_report(y_true, y_pred, zero_division=0))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    return acc, prec, rec, f1

def print_labels_with_zero_predictions(y_true, y_pred):
    """Print labels present in y_true but never predicted in y_pred."""
    unique_true = np.unique(y_true)
    unique_pred = np.unique(y_pred)
    missing = sorted(set(unique_true) - set(unique_pred))
    if missing:
        print("Labels present in test set with ZERO predictions:", missing)
    else:
        print("All labels in test set were predicted at least once.")

# ------------------------------------------------------------
# 5. RANDOM FOREST MODEL
# ------------------------------------------------------------
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

rf_scores = evaluate("Random Forest", y_test, rf_pred)
print_labels_with_zero_predictions(y_test, rf_pred)

# Feature Importance
plt.figure(figsize=(8,5))
sns.barplot(x=rf.feature_importances_, y=X.columns)
plt.title("Random Forest Feature Importance")
plt.show()

# ------------------------------------------------------------
# 6. BAGGING MODEL
# ------------------------------------------------------------
bag = BaggingClassifier(n_estimators=100, random_state=42)
bag.fit(X_train, y_train)
bag_pred = bag.predict(X_test)

bag_scores = evaluate("Bagging Classifier", y_test, bag_pred)
print_labels_with_zero_predictions(y_test, bag_pred)

# ------------------------------------------------------------
# 7. BOOSTING MODELS
# ------------------------------------------------------------
# AdaBoost - use SAMME to avoid FutureWarning
ada = AdaBoostClassifier(n_estimators=100, algorithm='SAMME', random_state=42)
ada.fit(X_train, y_train)
ada_pred = ada.predict(X_test)
ada_scores = evaluate("AdaBoost (SAMME)", y_test, ada_pred)
print_labels_with_zero_predictions(y_test, ada_pred)

# Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=200, random_state=42)
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
gb_scores = evaluate("Gradient Boosting", y_test, gb_pred)
print_labels_with_zero_predictions(y_test, gb_pred)

# ------------------------------------------------------------
# 8. MODEL COMPARISON SUMMARY
# ------------------------------------------------------------
summary = pd.DataFrame({
    "Model": ["Random Forest", "Bagging", "AdaBoost (SAMME)", "GradientBoosting"],
    "Accuracy": [rf_scores[0], bag_scores[0], ada_scores[0], gb_scores[0]],
    "Precision": [rf_scores[1], bag_scores[1], ada_scores[1], gb_scores[1]],
    "Recall": [rf_scores[2], bag_scores[2], ada_scores[2], gb_scores[2]],
    "F1-Score": [rf_scores[3], bag_scores[3], ada_scores[3], gb_scores[3]],
})

print("\n===== MODEL COMPARISON TABLE =====")
display(summary)
```
