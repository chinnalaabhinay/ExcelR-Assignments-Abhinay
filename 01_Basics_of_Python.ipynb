```python
#This program checks if the given number is prime

# Take user input as integer
num = int(input("Enter a number: "))

# Prime numbers are greater than 1
if num > 1:
    # Assume number is prime
    is_prime = True

    # Check from 2 to num - 1
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a prime number.")
    else:
        print(num, "is not a prime number.")
else:
    print(num, "is not a prime number.")
```

```python
import random

# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user for the product
print("What is the product of", num1, "and", num2, "?")
user_answer = int(input("Enter your answer: "))

# Calculate the correct answer
correct_answer = num1 * num2

# Check if the user's answer is correct
if user_answer == correct_answer:
    print(" Correct! Well done!")
else:
    print("Incorrect. The correct answer is", correct_answer)
```

```python
# Print squares of even numbers in range 100 to 200

print("Squares of even numbers from 100 to 200:")
for number in range(100, 201):
    # Check if number is even
    if number % 2 == 0:
        square = number * number
        print(number, "squared is", square)
```

```python
# Hospital Billing System Charges
room_per_day = 2000
doctor_fee = 1500
lab_test_fee = 300
discount_limit = 10000
discount_rate = 0.10

# Example input (can also use input())
days = 4
lab_tests = 3
medicine = 2400

# Calculate charges
room_total = days * room_per_day
lab_total = lab_tests * lab_test_fee
total_bill = room_total + doctor_fee + lab_total + medicine

# Discount calculation
if total_bill > discount_limit:
    discount = total_bill * discount_rate
else:
    discount = 0

final_amount = total_bill - discount

# Show bill
print("Room Charges:", room_total)
print("Doctor Fee:", doctor_fee)
print("Lab Tests:", lab_total)
print("Medicine Charges:", medicine)
print("Total:", total_bill)
print("Discount:", discount)
print("Final Bill:", final_amount)
```
