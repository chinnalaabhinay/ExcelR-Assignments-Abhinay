```python
#Task: Create a list of 5 integers and demonstrate: append, extend, insert, remove, pop, clear, index, count, sort, reverse.
def ex1_list_operations():
    nums = [5, 2, 9, 1, 7]
    print("Original:", nums)

    nums.append(11)
    print("append(11):", nums)

    nums.extend([3, 4])
    print("extend([3,4]):", nums)

    nums.insert(2, 0)   # insert 0 at index 2
    print("insert(2,0):", nums)

    nums.remove(9)      # remove first occurrence of 9
    print("remove(9):", nums)

    popped = nums.pop() # pop last
    print("pop():", nums, "popped:", popped)

    idx7 = nums.index(7) if 7 in nums else -1
    print("index(7):", idx7)

    cnt2 = nums.count(2)
    print("count(2):", cnt2)

    nums.sort()
    print("sort():", nums)

    nums.reverse()
    print("reverse():", nums)

    nums.clear()
    print("clear():", nums)

# Run example
ex1_list_operations()
```

```python
# Exercise 2: Tuple Example

# Creating a tuple with 3 student names
students = ("Ravi", "Anita", "Kiran")

# Trying to change the second name in the tuple
try:
    students[1] = "Sita"   # Tuples do not support item assignment
except TypeError as e:
    print("Error:", e)

# Printing the tuple to show it remains unchanged
print("Final tuple:", students)

# Explanation:
# Tuples in Python are immutable, meaning their elements cannot be changed,
# added, or removed after creation. Attempting to modify a tuple element
# raises a TypeError.
```

```python
#Grade from marks
#Task: Given marks, print grade:
#90+ → 'A'
#75–89 → 'B'
#50–74 → 'C'
#Below 50 → 'Fail'
# Exercise 6
def ex6_grade_from_marks(marks):
    print("Marks:", marks)
    if marks >= 90:
        print("Grade: A")
    elif marks >= 75:
        print("Grade: B")
    elif marks >= 50:
        print("Grade: C")
    else:
        print("Grade: Fail")

# Examples
ex6_grade_from_marks(92)
ex6_grade_from_marks(80)
ex6_grade_from_marks(60)
ex6_grade_from_marks(45)
```

```python
# Task: Given a number, check if it's positive, and if so check if it's even. If not positive, print zero or negative.
def ex7_positive_even_check(n):
    print("Number:", n)
    if n > 0:
        print("Positive number.")
        if n % 2 == 0:
            print("Also even.")
        else:
            print("Also odd.")
    elif n == 0:
        print("Zero.")
    else:
        print("Negative number.")

# Examples
ex7_positive_even_check(10)
ex7_positive_even_check(7)
ex7_positive_even_check(0)
ex7_positive_even_check(-5)
```

```python
#Pandas DataFrame (10 rows): .info(), .describe(), select/drop, fill missing Salary with mean, remove duplicates

#Task: Create a DataFrame with 10 rows columns: Name, Age, City, Salary. Then do the operations.
# Exercise 11
import pandas as pd
import numpy as np

data = {
    'Name': ['A','B','C','D','E','F','G','H','I','J'],
    'Age' : [25,30,22,28,35,30,22,28,29,35],
    'City': ['Hyd','Hyd','Bangalore','Mumbai','Delhi','Hyd','Bangalore','Mumbai','Delhi','Hyd'],
    # include some missing salaries and duplicates intentionally
    'Salary': [50000, 60000, None, 55000, 70000, 60000, None, 55000, 65000, 60000]
}
df = pd.DataFrame(data)
print("Initial DataFrame:\n", df, "\n")

# .info() and .describe()
print("Info:")
df.info()
print("\nDescribe:")
print(df.describe(include='all'))

# Select Name and City
print("\nSelected columns (Name, City):\n", df[['Name','City']])

# Drop City
df_no_city = df.drop(columns=['City'])
print("\nAfter dropping City:\n", df_no_city)

# Fill missing Salary with mean
salary_mean = df['Salary'].mean(skipna=True)
df['Salary'] = df['Salary'].fillna(salary_mean)
print(f"\nAfter filling missing Salary with mean ({salary_mean:.2f}):\n", df)

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nAfter removing duplicates:\n", df_no_duplicates)
```
